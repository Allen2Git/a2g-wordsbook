```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
# @Time        : 2020, 04, 14, 05, 41
# @Author      : Allen Zhang
# @File        : ~/zhangyang/workspace/a2g-wordsbook/main/media_json_confidence.ipynb
# @Description : Analysis the media json file from AWS Transcribe

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import json
media_json_file = open("/Users/zyaws/zhangyang/workspace/a2g-wordsbook/main/words/media-overview.json")
media_json = json.loads(media_json_file.read())
media_confidence = []
for item in media_json["results"]["items"]:
    media_confidence.append(float(item["alternatives"][0]["confidence"]))
```

```python
x_axis = range(0,len(media_confidence))
y_axis = sorted(media_confidence)

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.scatter(x_axis,y_axis)
plt.show()

# sorted 2876 element is 0.9 percentile
```

```python
i=0
for num in y_axis:
    i += 1
    if num>0.9:
        print(i)
        break
```
